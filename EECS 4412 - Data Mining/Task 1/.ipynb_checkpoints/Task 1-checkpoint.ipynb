```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
```

```python
df = pd.read_csv('200_selection.csv', header = 0)
df2 = pd.read_csv("test2", header = 0)
```

```python
selected_columns = df[['Home Team Possession %', 'Away Team Possession %', 'Home Team Pass Success %', 'Away Team Pass Success %', 'Home Team Total Shots', 'Away Team Total Shots', 'Home Team Blocked Shots', 'Away Team Blocked Shots', 'Result']]
```

Currently set as 200 A wins, 200 B wins, 200 ties, possibly choose different cats or maybe only show a or b wins
Green in middle because ties, blue and red more seperated

```python
#set figure size
plt.figure(figsize=(18,8))

#plot parallel coordinates graph 
parallel_coordinates(selected_columns, "Result", color=['red', 'blue', 'green'])
#plt.show()
```

```python
selected_columns2 = df2[['Home Team Possession %', 'Home Team Pass Success %', 'Home Team Total Shots', 'Home Team Blocked Shots', 'Home Team Rating', 'league']]
```

```python
#set figure size
plt.figure(figsize=(18,8))

#plot parallel coordinates graph 
parallel_coordinates(selected_columns2, "league", color=['red', 'blue', 'green', 'violet', 'yellow'])
plt.show()
```
